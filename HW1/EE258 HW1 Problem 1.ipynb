```python
import numpy as np
import matplotlib.pyplot as plt
import math


z = np.linspace(-5,5)

phi_1 = 1 / (1+np.exp(-z))

fig = plt.figure()
plt.grid()
plt.plot(z,phi_1,'r')

phi_2 = np.tanh(z)
plt.plot(z,phi_2,'b')

def relu(x):
    return np.maximum(0,x)

phi_3 = relu(z)
plt.plot(z,phi_3,'g')
```
